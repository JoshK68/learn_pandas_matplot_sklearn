#Matplot Tutorial


```python
import matplotlib.pyplot as plt
```

```python
plt.plot() #There's no data, so it's an empty plot...Simply shows the X and y
```

```python
###What do you pass into plt.plot()?
#it expects and x or a y or both in 'array-like format' such as a numpy array
#you can also pass in other parameters
```

```python
###If we only pass in one value, then it will assume that they are y values
#and it will create x values starting at 0 and incrementing by 1

plt.plot([2, 4, 6, 8])  #Matplot assumes that the values in this array are y values
#So it's making:
#  y =[2, 4, 6, 8]
#  and
#  x = [0, 1, 2, 3]
```

```python
#We can give it more interesting numbers

plt.plot([2, 4, 2, 6, 8, 4])   #and it makes up it's own x axis
```

```python
#We can also pass in an x. In that case it will assume the first list that we pass in is the X

#We can use keyword arguments to make it less confusing

salaries = [55000, 65000, 72000, 90000, 115000, 150000]
ages = [20, 25, 30, 32, 40, 45]

#let's put ages on the X axis

plt.plot(ages, salaries)
```

```python
import numpy as np
```

```python
#Let's use numpy to quickly create a bunch of numbers
nums = np.arange(5)  #stands for 'a range'
#This simply gives us a list of nums from 0 up to 5
#but what's most important is that we can do math with it
#ex:
nums * nums
```

```python
plt.plot(nums, nums)
plt.plot(nums, nums*nums)  #adds a second line to the same graph
plt.show()
```

```python
###Think of a figure as a blank visual canvas
###Each individual plot consists of an axis. 
###2D plots have x and y. Can also be 3D+ plots
###Must watch this again! https://www.udemy.com/course/python-data-analysis-visualization/learn/lecture/29613926#overview
###There can be multiple axis (plots) in a figure

###We can totally modify this so that each line is on its own axis
###Or we can break the above plot into 2 figures each with it's own axis (plot)
### To make these layout changes, we use pyplot.figure()
```

#Figure vs Plot


```python
plt.figure()  ###Everything that comes after is added to this figure!
plt.plot(nums, nums)
plt.figure()  ###B/c we ran figure twice, we get two separate figures!
plt.plot(nums, nums*nums)  #adds a second line to the same graph
plt.show()
```

```python
plt.figure()
plt.plot(nums, nums)
plt.figure()
plt.plot(nums, nums*nums)
plt.figure()
plt.plot(nums, nums**3)
plt.show()
```

```python
plants = ['Spinach', 'Turnip', 'Rhubarb', 'Broccoli', 'Kale']
died = [10, 25, 5, 30, 21]
germinated = [74, 88, 56, 69, 59]
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, germinated, width=0.5, color='b');  #Remember x goes before y in both the alphabet and here
###To make it a bar chart all you need to do is plt.bar!
#You can give it a title
plt.title('Germination Rates');
#You can also give it a ylabel
plt.ylabel('Number of seedlings')
```

```python
plt.figure(figsize=(6, 6))
plt.bar(plants, died, color='r') #instead of plt.plot   it's plt.bar
plt.title('Failure Rates')
plt.ylabel('Number of Dead Seedlings')
```

```python
#We can easily plot both together

plt.figure(figsize=(6, 6))
plt.bar(plants, died, color='r', width=0.4, align='edge')
plt.bar(plants, germinated, width=0.4);

###change the width and change the color, and even align at the edge, so that they don't overlap... otherwise they'd totally overlap
```

```python
#Stacked plots are also very cool, and all you have to do is add 'bottom'

plt.figure(figsize=(6, 6))
plt.bar(plants, died, color='r', width=0.4, label='Failed Seedlings')
plt.bar(plants, germinated, width=0.4, bottom=died, label='Germinated Seedlings');
plt.legend();           ###B/c you added labels, you also need to add plt.legend() to show them
###bottom allows you to very easily see that far more germinated than died
```
